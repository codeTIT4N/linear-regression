```python
#Importing libraries
import pandas
from sklearn import linear_model
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
# Getting data from csv file
data = pandas.read_csv('company_data.csv')
print(['TV', 'Radio', 'Newspaper'], " are the feature columns, " + "whereas 'Sales' is the response column.\n")
```

```python
# Plotting data
plt.title('Plot of Data')
plt.xlabel('Advertising Expense')
plt.ylabel('Sales')
plt.scatter(data['TV'], data['Sales'], label = "TV")
plt.scatter(data['Radio'], data['Sales'], label = "Radio")
plt.scatter(data['Newspaper'], data['Sales'], label = "Newspaper")
plt.legend(ncol=3, loc="lower right")
plt.show()
```

```python
# These are the feature columns
features = data[['TV', 'Radio', 'Newspaper']]
# These are the response columns
response = data['Sales']
# We now split the dataset into training and testing datasets
features_train, features_test, response_train, response_test = train_test_split(features, response, test_size=0.2)
```

```python
# Here we perform the regression
regression = linear_model.LinearRegression()
model = regression.fit(features, response)
print("\nRegression Coefficients/Beta Values: ", model.coef_)
print("\nRegression Intercept/Alpha: ", model.intercept_)
new_feature_data = [[152, 11, 19]]
print("\nPredicting salse for TV, Radio, Newspaper advertisement expenditure of ",new_feature_data, " we get: ", model.predict(new_feature_data))
```
